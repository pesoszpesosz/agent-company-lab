## Synthetic Smoke Test

This optional toy check uses only inline synthetic rows. It validates notebook structure without touching any competition data.

```python
toy_rows = [
    {"feature_a": 0.1, "feature_b": 1, "target": 0},
    {"feature_a": 0.4, "feature_b": 0, "target": 0},
    {"feature_a": 0.8, "feature_b": 1, "target": 1},
    {"feature_a": 0.9, "feature_b": 1, "target": 1},
]

def majority_class(rows, target="target"):
    counts = {}
    for row in rows:
        counts[row[target]] = counts.get(row[target], 0) + 1
    return max(counts, key=counts.get)

def accuracy(rows, prediction, target="target"):
    return sum(1 for row in rows if row[target] == prediction) / len(rows)

toy_prediction = majority_class(toy_rows)
toy_accuracy = accuracy(toy_rows, toy_prediction)
{"toy_prediction": toy_prediction, "toy_accuracy": toy_accuracy}
```
